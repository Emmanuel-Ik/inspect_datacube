```python


from datacube import Datacube
from datetime import datetime

```

```python
# Connect to the datacube
dc = Datacube()

```

```python
# Specify the product name
product_name = 'sen2_l2a_gcp'

```

```python
dc.list_products()
```

```python
dc.list_measurements()
```

```python
# Get the product definition
product = dc.index.products.get_by_name(product_name)


product

```

```python
if product:
    # Get the available time range for the product
    time_extents = dc.index.products.get_product_extents(product_name)
    
    # Extract start and end dates
    start_date = time_extents['times'][0].strftime('%Y-%m-%d')
    end_date = time_extents['times'][1].strftime('%Y-%m-%d')

    print(f"Start Date: {start_date}")
    print(f"End Date: {end_date}")
else:
    print(f"Product '{product_name}' not found.")
```

```python
# for [time] in dc.index.datasets.search_returning(['time'], product=product_name):
#    print(time)
#    # Time is actually a range
# start_time, end_time = time.lower, time.upper

all_time = [i for i in dc.index.datasets.search_returning(['time'], product=product_name)]

# len(all_time) 18648
# type(all_time[0]) datacube.index.postgres._datasets.search_result

all_time[0]
```

```python

min_date = all_time[0].time.lower
max_date = all_time[0].time.upper


for result in all_time[0:]:
    if result.time.lower < min_date:
        min_date = result.time.lower
    if result.time.upper > max_date:
        max_date = result.time.upper
 
min_date = min_date.strftime('%Y-%m-%d %H:%M:%S')
max_date = max_date.strftime('%Y-%m-%d %H:%M:%S')

print(f"Minimum Date: {min_date}")
print(f"Maximum Date: {max_date}")
```
